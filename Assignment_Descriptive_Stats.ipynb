**Question 1. Understanding Central Tendency**

**A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].**

**What is the most representative value of their weekly sales, and why?**
- Understanding Central Tendency

Given daily muffin sales (in dozens):
10, 12, 11, 15, 14, 13, 12

Mean = 87 / 7 ≈ 12.43

Median = 12

Mode = 12

Hence, The most representative value is 12 dozens because it is both the median and the mode, and it best reflects the typical daily sales without being influenced by extreme values.

Why is the median a better measure than the mean in this case? Calculate the median**

- Median in Real Life

First, arrange the car prices in ascending order:
$8,000, $9,500, $10,200, $11,000, $50,000

Median Calculation

The median is the middle value, which is $10,200.

Thus, The median is a better measure than the mean because the very high price of $50,000 is an outlier that would increase the average and give a misleading picture of typical car prices. The median represents the true central value of the dataset.
